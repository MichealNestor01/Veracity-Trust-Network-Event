```python
# 3.1
example_data['VisitTimeFlag'] = 0
example_data['PagesViewedFlag'] = 0
example_data['ActionsPerformedFlag'] = 0
example_data['BrowserFlag'] = 0
example_data['MaliciousBotPrediction'] = 0
```

```python
# 3.2
for index, row in example_data.iterrows():
    #data_df.at[index, 'announcedBot'] = 1
    if row['totalVisitTime'] == 0:
        example_data.at[index, 'VisitTimeFlag'] = 1
    if row['pagesViewedInVisit'] == 1:
        example_data.at[index, 'PagesViewedFlag'] = 1
    if row['totalActionsCount'] == 0:
        example_data.at[index, 'ActionsPerformedFlag'] = 1
    if row['browser'] not in ["Chrome", "Safari", "Firefox", "Chromium", "Netscape", "Microsoft Internet Explorer", "Google Chrome"]:
        example_data.at[index, 'BrowserFlag'] = 1
```

```python
# 3.3
for index, row in example_data.iterrows():
    if row['announcedBot'] == 1:
        example_data.at[index, 'MaliciousBotPrediction'] = 0
    else:
        flag_count = row['VisitTimeFlag'] + row['PagesViewedFlag'] + row['ActionsPerformedFlag'] + row['BrowserFlag']
        example_data.at[index, 'MaliciousBotPrediction'] = flag_count * 25

```
